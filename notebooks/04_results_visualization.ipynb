```python
# ── Figure 10: ROC Curves (simulated) ────────────────────────────────
# Simulate scores for illustration (replace with real predictions after training)
np.random.seed(42)
N = 1000
y_true = np.random.randint(0, 2, N)
def sim_scores(auc_target, n=N):
    signal = y_true * auc_target + np.random.randn(n) * (1 - auc_target) * 0.5
    return (signal - signal.min()) / (signal.max() - signal.min())

roc_scores = {
    'SVM': sim_scores(0.86),
    'Random Forest': sim_scores(0.89),
    'LSTM': sim_scores(0.925),
    'GNN': sim_scores(0.936),
    'GT-ADF (Proposed)': sim_scores(0.97),
}
plot_roc_curves(y_true, roc_scores, save_path='../results/figures/fig10_roc.png')
```
